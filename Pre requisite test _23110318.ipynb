### Q.2 Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

```python
# Calculating time in matrix multiplication using list of lists
import time

start_time = time.time()
def mul(a, b):
    c = []
    for i in range(len(a)):
        c.append(a[i] * b[i])
    return sum(c)
    
def matmul(a,b):
    c = [[row[i] for row in b] for i in range(len(a))]
    d = []
    for i in range(len(a)):
        e = []
        for j in range(len(c)):
            e.append(mul(a[i], c[j]))
        d.append(e)
    return d

a = eval(input())
b = eval(input())
print(matmul(a, b))
end_time = time.time()
print("Time in matrix multiplication using list of lists: {}".format(end_time - start_time))
```

### Q.5 Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
import jax
import jax.numpy as jnp

def f(x, y):
    return ((x**2) * y) + ((y**3) * jnp.sin(x))

# differentiating wrt x
grad_f_x = jax.grad(f, argnums = 0)

# differnetiating wet y
grad_f_y = jax.grad(f, argnums = 1)

print(grad_f_x(float(0),float(0)))
print(grad_f_x(float(1),float(0)))
print(grad_f_x(float(0),float(1)))
print(grad_f_x(float(1),float(1)))
print(grad_f_y(float(0),float(0)))
print(grad_f_y(float(1),float(0)))
print(grad_f_y(float(0),float(1)))
print(grad_f_y(float(1),float(1)))
```

### Q.6 Use sympy to confirm that you obtain the same gradient analytically.

```python
import sympy as sp

x, y = sp.symbols("x y")

f = ((x**2) * y) + ((y**3) * sp.sin(x))

gradient = sp.Matrix([f.diff(x), f.diff(y)])

gradient_value = gradient.subs([(x, 1), (y, 1)])

print(gradient)
print(gradient_value)
```

### Q.10 Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. ### Q.11 Add a new column to the dataframe called f which is the sum of the columns a, b, c, d, e. Create another column called g. The value in the column g should be LT8 if the value in the column f is less than 8 and GT8 otherwise. Find the number of rows in the dataframe where the value in the column g is LT8. Find the standard deviation of the column f for the rows where the value in the column g is LT8 and GT8 respectively.

```python
# adding a coulmn f which is the sum of all columns
df["f"] = df["a"] + df["b"] + df["c"] + df["d"] + df["e"]

# adding column g
def compare(value):
    if value < 8:
        return "LT8"
    else:
        return "GT8"
df["g"] = df["f"].apply(compare)
 
# finding the number of rows as asked
print("The number of rows where the value in the column g is LT8 is {}".format((df["g"] == "LT8").sum()))
print("The number of rows where the value in the column g is GT8 is {}".format((df["g"] == "GT8").sum()))

# finding the standard deviations
print("The std of column 'f' for row in column 'g' is 'LT8' is: {}".format(df[df["g"] == "LT8"]["f"].std()))
print("The std of column 'f' for row in column 'g' is 'GT8' is: {}".format(df[df["g"] == "GT8"]["f"].std()))
```
